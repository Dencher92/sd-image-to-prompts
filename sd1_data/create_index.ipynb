```python
import os
from tqdm import tqdm
import json
import pandas as pd
```

```python
imgs_root = '/mnt/home/data/diffusiondb_img'
```

```python
index = []
for part in tqdm(os.listdir(imgs_root)):
    part_meta_path = os.path.join(imgs_root, part, f'{part}.json')
    with open(part_meta_path, 'r') as meta_fobj:
        meta = json.load(meta_fobj)
        for img_name in meta.keys():
            prompt = meta[img_name]['p']
            diffusion_steps = meta[img_name]['st']
            index.append([part, img_name, prompt, diffusion_steps])
```

```python
df = pd.DataFrame(index, columns=['part', 'img_name', 'prompt', 'steps'])
df['part_id'] = df.groupby('part').ngroup()
df = df.sort_values('part_id')
df = df.reset_index(drop=True)
```

```python
df['image_number'] = df.index
```
